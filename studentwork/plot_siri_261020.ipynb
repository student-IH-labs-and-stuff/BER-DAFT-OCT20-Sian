```python
# hypothesis : MPG is ‘driven’ by horesepower
# and is not affected by # cylinders
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('mtcars.csv')
data.head()
```

```python
data.dtypes
```

```python
# H1

y = data['MPG']
x1 = data['Horsepower']
x2 = data['Cylinders']
```

```python
sns.regplot(x1, y)
```

```python
sns.boxplot(x2, y)
```

```python
correlations_matrix = data.corr()
sns.heatmap(correlations_matrix, annot = True)
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize = (10, 8))
ax = sns.heatmap(correlations_matrix, mask = mask, annot = True)
plt.show()
```

```python
# We will reject the hypothesis since the number of cylinders also
# has a negative correlation to MPG

```
